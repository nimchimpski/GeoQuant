```python
import numpy as np

class BayesianTradeDecision:
    def __init__(self):
        self.prior_probs = np.array([0.45, 0.25, 0.30])  # drop, bounce, flat
    
    def update(self, is_positive_news=False, evidence_strength=0.7):
        lik = np.array([0.65, 0.15, 0.20]) if not is_positive_news else np.array([0.25, 0.55, 0.20])
        post = self.prior_probs * lik
        post /= post.sum()
        return post
    
    def decide(self, post, cash_need=0.30):
        p_drop = post[0]
        ev = p_drop * (-0.08) + (1-p_drop) * 0.03
        if p_drop > 0.40 and ev < -0.02:
            return f"SELL {cash_need*100:.0f}% NOW", ev
        return "HOLD", ev

model = BayesianTradeDecision()
post = model.update(is_positive_news=False)          # ← 11am CET reality
decision, ev = model.decide(post)
print(f"Decision: {decision} | Expected return if hold: {ev*100:+.1f}%")
```
