```python
def plotMeans(numDice, numRolls, numBins, legend, colour, style):
    means = []
    print(f"Rolling {numDice} dice {numRolls//numDice} times")
    for i in range(numRolls):
        print(f'roll {i}')
        vals = 0
        for j in range(numDice):
            roll = 5 * random.random()
            print(f' die {j} rolls {roll}')
            # progreesively add a random number between 1 and 5
            vals += roll
        print(f'mean for this roll {vals/float(numDice)}')
        # get the average and append to means
        means.append(vals/float(numDice))
    plt.hist(means, bins=numBins, density=True, histtype=style, color=colour, alpha=0.5, label=legend)
    #  get means and std of means
    mean = sum(means)/len(means)
    std = (sum([(x-mean)**2 for x in means])/len(means))**0.5
    return (mean, std)

mean, std = plotMeans(10, 100000, 19, '1 die', 'blue', 'stepfilled')
print("1 die: mean = %.2f, std = %.2f" % (mean, std))
# mean, std = plotMeans(50, 100000, 19, '50 dice', 'red', 'stepfilled')
# print("50 dice: mean = %.2f, std = %.2f" % (mean, std))

```
