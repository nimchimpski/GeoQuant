```python
import numpy as np
import pandas as pd

a=np.array([2,3,4,5,np.nan, 6])
b=np.array([6,7,8,9,10,11])
c=np.array([1,2,3,4,5,6,np.nan,8,9,])
s1 = pd.Series(a)
s2=pd.Series(b)
s3=pd.Series(c)

df = pd.DataFrame({'s1': s1, 's2':s2, 's3': s3})
print(df)
print('')

df=df.dropna(how='all')
# df['s3']=df['s3'].shift(-1)
print(df)

```
