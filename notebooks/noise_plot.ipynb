```python
import matplotlib.pyplot as plt
# dark theme
plt.style.use('dark_background')
labels = [
    f"Geometric\nmisregistration /\nlayover /\nshadow",
    "Temporal\ndecorrelation",
    "Speckle\nvariance",
    "Incidence-\nangle \nvariation",
    f"Atmospheric path /\nionosphere",
    "Metadata /\ngeocoding errors",
    "Radiometric\ndrift"
]
values = [25, 20, 15, 12, 10, 10, 8]

# plot donut chart
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))
wedges, texts, autotexts = ax.pie(
    values, labels=labels, autopct='%2.1f%%', startangle=140,
    wedgeprops=dict(width=0.6), pctdistance=0.8
)
plt.setp(autotexts, size=11, weight="bold", color="black")
ax.set_title(f"Breakdown of Signal Corruption Sources\nin Urban SAR Workflows (illustrative)", fontsize=16)
ax.text(0,0,"only ≈\n 50% signal\nremains.", ha='center', va='center', fontsize=14, color='white') 
plt.show()                                                    
```
